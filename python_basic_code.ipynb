# List processing in python
steps:
step 1: create a empty list
step2:keep the total variable string at 0
sep3:loop through each number
step4:add to total ,append total to result

```python
def runningSum(nums):
  result=[]#Empty list
  total=0 #Initialise total to zero
  for n in nums:#stat the loop for running sum
    total += n#add the value to total
    result.append(total)#append the sum to result list
  return(result)
print(runningSum([10,-20,5,1]))
```

```python
def runningSum(nums):
  total=0
  for n in nums:
    total += n
  return(total)
print(runningSum([10,20,5,1]))
```

```python
def runningSum(nums):
  count=0
  for n in nums:
    if n%2==0:
      count+=1
  return(count)
print(runningSum([10,20,52,1]))
```

```python
def reverselist(nums):
  return nums[::-1]
print(reverselist([1,2,3,4]))
my_list = [10,5,1,0,-67]
print(my_list[::2])#the list slices from the beginning to end, taking every second element
print(my_list[::3])
print(my_list[::-2])
print(my_list[::-3])
```

# :: in Python enables flexibles extraction and manipuation of sequence elements.

```python
#Find maximum number in the list
def FindMax(nums):
  max_num = nums[0]
  for n in nums:
    if n>max_num:
      max_num = n
  return max_num
print(FindMax([3,7,2,9,5]))
```

```python
#Find minimum number in the list
def FindMin(nums):
  min_num = nums[0]
  for n in nums:
    if n<min_num:
      min_num = n
  return min_num
print(FindMin([3,7,2,9,5]))
```
