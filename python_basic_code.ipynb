# List processing in python
steps:
step 1: create a empty list
step2:keep the total variable string at 0
sep3:loop through each number
step4:add to total ,append total to result

```python
def runningSum(nums):
  result=[]#Empty list
  total=0 #Initialise total to zero
  for n in nums:#stat the loop for running sum
    total += n#add the value to total
    result.append(total)#append the sum to result list
  return(result)
print(runningSum([10,-20,5,1]))
```

```python
def runningSum(nums):
  total=0
  for n in nums:
    total += n
  return(total)
print(runningSum([10,20,5,1]))
```

```python
def runningSum(nums):
  count=0
  for n in nums:
    if n%2==0:
      count+=1
  return(count)
print(runningSum([10,20,52,1]))
```

```python
#Find maximum number in the list
def FindMax(nums):
  max_num = nums[0]
  for n in nums:
    if n>max_num:
      max_num = n
  return max_num
print(FindMax([3,7,2,9,5]))
```

```python
#Find minimum number in the list
def FindMin(nums):
  min_num = nums[0]
  for n in nums:
    if n<min_num:
      min_num = n
  return min_num
print(FindMin([3,7,2,9,5]))
```

## grocery store

```python
# Product list and prices
products = ["Vegetables", "chocolate", "milk", "dairy item"]
prices = [200, 120, 50, 60]

# Empty cart
cart = []

# Show available products
def show_products():
    print("\nAvailable Products:")
    for i in range(len(products)):
        print(f"{i + 1}. {products[i]} - ₹{prices[i]}")

# Add item to cart
def add_to_cart():
    show_products()
    choice = int(input("Enter product number to add: ")) - 1
    if choice >= 0 and choice < len(products):
        cart.append(choice)
        print(products[choice], "added to cart.")
    else:
        print("Invalid choice.")

# Remove item from cart
def remove_from_cart():
    if len(cart) == 0:
        print("Cart is empty.")
        return
    print("\nItems in Cart:")
    for i in range(len(cart)):
        print(f"{i + 1}. {products[cart[i]]} - ₹{prices[cart[i]]}")
    choice = int(input("Enter item number to remove: ")) - 1
    if choice >= 0 and choice < len(cart):
        removed = cart.pop(choice)
        print(products[removed], "removed from cart.")
    else:
        print("Invalid choice.")

# View cart
def view_cart():
    if len(cart) == 0:
        print("Cart is empty.")
        return
    print("\nYour Cart:")
    total = 0
    for i in range(len(cart)):
        item = cart[i]
        print(f"{i + 1}. {products[item]} - ₹{prices[item]}")
        total += prices[item]
    print("Total: ", total)

# Checkout
def checkout():
    view_cart()
    if len(cart) > 0:
        print("Thank you for shopping!")
    else:
        print("Cannot checkout. Cart is empty.")
    print("Exiting...")
    exit()

while True:
    print("\n=== Grocery Store Menu ===")
    print("1. Show Products")
    print("2. Add to Cart")
    print("3. Remove from Cart")
    print("4. View Cart")
    print("5. Checkout and Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        show_products()
    elif choice == '2':
        add_to_cart()
    elif choice == '3':
        remove_from_cart()
    elif choice == '4':
        view_cart()
    elif choice == '5':
        checkout()
    else:
        print("Invalid choice. Try again.")
```

```python

class CoffeeMachine:
    def __init__(self):
      self.water = 100

    def make_coffee(self):
      if self.water >= 10:
        self.water -= 10
        print("Coffee is ready!")
      else:
        print('Not enough water.')

machine = CoffeeMachine()
machine.make_coffee()
```

```python
class BankAccount:
  def __init__(self):
    self.__balance = 0
  def set_balance(self, amount):
    self.__balance = amount
  def get_balance(self):
    return self.__balance

account = BankAccount()
account.set_balance(1000)
print("Account balance =", account.get_balance())
```

```python
class animal:
    def move(self):
        print("animal is moving")

class dog(animal):
    def bark(self):
        print("dog is barking")

class cat(animal):
    def meow(self):
        print("cat is meowing")

dog = dog()
dog.move()
dog.bark()

cat = cat()
cat.move()
cat.meow()
```
