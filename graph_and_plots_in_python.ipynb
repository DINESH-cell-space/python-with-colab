```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# User input
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")
a = float(input("Enter coefficient a (for ax^3): "))
b = float(input("Enter coefficient b (for bx^2): "))
c = float(input("Enter coefficient c (for cx): "))
d = float(input("Enter constant d: "))

# Define symbolic variable and function
x = sp.Symbol('x')
f = a*x**3 + b*x**2 + c*x + d

# First derivative
f_prime = sp.diff(f, x)
print("First Derivative f'(x):", f_prime)

# Critical points
cp = sp.solve(f_prime, x)
print("Critical Points:", cp)

# Second derivative
f_double_prime = sp.diff(f, x, 2)
print("Second Derivative f''(x):", f_double_prime)

# Classification of critical points
for point in cp:
    concavity = f_double_prime.subs(x, point)
    nature = "Minimum" if concavity > 0 else "Maximum" if concavity < 0 else "Inflection Point"
    print(f"At x = {point}, f''(x) = {concavity}  {nature}")
```
