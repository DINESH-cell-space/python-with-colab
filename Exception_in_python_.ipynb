```python
print("Demonstration of Exception in python\n")

# 1. Zero division error
try:
  print(10/0)
except ZeroDivisionError:
  print("(X) Zero division error: You cannot divided by zero \n")

# 2. value error
try:
  num = int("abc")
except ValueError:
  print("(X) Value error: cannot convert string to integer \n")

# 3. type error
try:
  result = "2" + 2
except TypeError:
  print("(X) Type error: you tried to add a string and integer \n")

# 4. index error
try:
  lst = [1, 2, 3]
  print(lst[5])
except IndexError:
  print("(X) Index Error: list index out of range \n")

#5 Key Error
try:
  my_dict = {"name":'saravana'}
  print(my_dict["c"])
except KeyError:
  print("(X) key Error: not found in dictionary")

#6 file not found error
try:
  open("missing.txt")
except FileNotFoundError:
  print("(x) File not found error: file not found  \n")

# 7. name error
try:
  print("undefined varialble")
except NameError:
  print("(X) Name error: Variable is not found")
```
