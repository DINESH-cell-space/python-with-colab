```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import requests
import time
import sqlite3
import json
from datetime import datetime
import tkinter as tk
from tkinter import ttk, messagebox, scrolledtext
from tkinter.font import Font

# Code from cell lM_-cu5Fu4wt (Bar Graph Representation)
# create a nested dictionary with persons data
person={
   "p1":{"SleepHours":7.5,"DeepSleep(hrs)":2.0,"LightSleep(hrs)":5.0,"Wakeups":1},
    "p2":{"SleepHours":8.0,"DeepSleep(hrs)":2.5,"LightSleep(hrs)":5.0,"Wakeups":0},
    "p3":{"SleepHours":5.5,"DeepSleep(hrs)":1.0,"LightSleep(hrs)":3.8,"Wakeups":3},
 "p4":{"SleepHours":6.8,"DeepSleep(hrs)":1.6,"LightSleep(hrs)":4.7,"Wakeups":2}

}
data_person=pd.DataFrame(person).T # Renamed 'data' to 'data_person' to avoid clash with string 'data'

data_person["Average Sleep Time"]=data_person.mean(axis=1)

# Bar chart for person's average sleep time
plt.figure(figsize=(8,5))
data_person["Average Sleep Time"].plot(kind="bar",color="purple",edgecolor="black")
plt.title("Average Sleep Time of Person")
plt.xlabel("Person")
plt.ylabel("Average sleep time in hours")
plt.show()

# Code from cell CLuxFf-Vvgv1 (Pie Chart Representation)
student =["p1","p2","p3", "p4"]
Fees =[2.0,2.5,1.0,1.6]
plt.pie(Fees, labels=student, autopct='%1.1f%%', startangle=90)
plt.title("Sleep Distribution")
plt.show()

# Code from cell hg28cLnFh7pK (Read File) - Removed file reading due to FileNotFoundError
# The content of the file is available as a string in data_sleep_quality below.

# Code from cell -Dq40OOCj8eV (sleep quality analysis)
data_sleep_quality = """
Person,SleepHours,DeepSleep(hrs),LightSleep(hrs),Wakeups
P1,7.5,2.0,5.0,1
P2,8.0,2.5,5.0,0
P3,5.5,1.0,3.8,3
P4,6.8,1.6,4.7,2
P5,7.9,2.3,5.2,0
P6,6.5,1.4,4.6,2
P7,8.2,2.6,5.3,0
P8,7.0,1.8,4.9,1
P9,6.7,1.6,4.5,2
P10,7.4,2.1,5.0,1
"""

df = pd.read_csv(StringIO(data_sleep_quality))

def compute_sleep_quality(row):
    deep_ratio = row['DeepSleep(hrs)'] / row['SleepHours']
    light_ratio = row['LightSleep(hrs)'] / row['SleepHours']
    wake_penalty = 1 - (row['Wakeups'] / 5)
    score = 50 * deep_ratio + 30 * light_ratio + 20 * wake_penalty
    return round(score, 2)

df['SleepQualityScore'] = df.apply(compute_sleep_quality, axis=1)

plt.figure(figsize=(10,6))
plt.bar(df['Person'], df['SleepQualityScore'], color='skyblue')
plt.title('Sleep Quality Score per Person')
plt.xlabel('Person')
plt.ylabel('Sleep Quality Score (0‚Äì100)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

display(df[['Person', 'SleepHours', 'DeepSleep(hrs)', 'LightSleep(hrs)', 'Wakeups', 'SleepQualityScore']])

# Code from cell kcONRWsCkj-1 (sleep quality analysis with rank)
# The compute_sleep_quality function is already defined above, no need to redefine
df['SleepQualityScore'] = df.apply(compute_sleep_quality, axis=1)
df['Rank'] = df['SleepQualityScore'].rank(ascending=False, method='min').astype(int)
df = df.sort_values(by='Rank')

plt.figure(figsize=(10,6))
plt.bar(df['Person'], df['SleepQualityScore'], color='skyblue')
plt.title('Sleep Quality Score per Person')
plt.xlabel('Person')
plt.ylabel('Sleep Quality Score (0‚Äì100)')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

display(df[['Rank', 'Person', 'SleepHours', 'DeepSleep(hrs)', 'LightSleep(hrs)', 'Wakeups', 'SleepQualityScore']])

# Code from cell k3rrbIikHLp7 (PubMed Chatbot App)
# All imports already handled at the top

def init_db():
    conn = sqlite3.connect("chatbot_history.db")
    cur = conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS history(id INTEGER PRIMARY KEY AUTOINCREMENT, user_id INTEGER, query TEXT, timestamp TEXT, results TEXT)")
    cur.execute("CREATE TABLE IF NOT EXISTS users(id INTEGER PRIMARY KEY AUTOINCREMENT, username TEXT UNIQUE, password TEXT)")
    try:
        cur.execute("INSERT INTO users(username, password) VALUES (?, ?)", ("demo", "demo123"))
        conn.commit()
    except sqlite3.IntegrityError:
        pass
    return conn


def authenticate_user(conn, username, password):
    cur = conn.cursor()
    cur.execute("SELECT id FROM users WHERE username=? AND password=?", (username, password))
    result = cur.fetchone()
    return result[0] if result else None


def save_history(conn, user_id, query, results):
    cur = conn.cursor()
    cur.execute("INSERT INTO history(user_id, query, timestamp, results) VALUES (?, ?, ?, ?)", (user_id, query, datetime.now().isoformat(), json.dumps(results)))
    conn.commit()


def get_history(conn, user_id, limit=10):
    cur = conn.cursor()
    cur.execute("SELECT query, timestamp FROM history WHERE user_id=? ORDER BY id DESC LIMIT ?", (user_id, limit))
    return cur.fetchall()


def fetch_summary(pubmed_id):
    url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi"
    params = {"db": "pubmed", "id": pubmed_id, "retmode": "json"}
    try:
        r = requests.get(url, params=params, timeout=10)
        r.raise_for_status()
        d = r.json()["result"].get(pubmed_id, {})
        return {"title": d.get("title", "No title"), "journal": d.get("fulljournalname", "Unknown"), "authors": ", ".join([a.get("name", "") for a in d.get("authors", [])][:3]), "pub_date": d.get("pubdate", "N/A"), "url": f"https://pubmed.ncbi.nlm.nih.gov/{pubmed_id}/"}
    except:
        return None


def search_pubmed(term, max_results=5):
    base = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi"
    params = {"db": "pubmed", "term": term, "retmode": "json", "retmax": max_results}
    try:
        r = requests.get(base, params=params, timeout=10)
        r.raise_for_status()
        ids = r.json().get("esearchresult", {}).get("idlist", [])
        papers = []
        for pid in ids:
            data_summary = fetch_summary(pid) # Renamed 'data' to 'data_summary'
            if data_summary:
                papers.append(data_summary)
            time.sleep(0.3)
        return papers
    except:
        return []


class PubMedChatbotApp:
    def __init__(self, root):
        self.root = root
        self.root.title("AI Health Chatbot - PubMed Powered")
        self.root.geometry("900x650")
        self.root.configure(bg="#f0f4f8")
        self.conn = init_db()
        self.user_id = None
        self.username = None
        self.title_font = Font(family="Helvetica", size=24, weight="bold")
        self.heading_font = Font(family="Helvetica", size=14, weight="bold")
        self.normal_font = Font(family="Helvetica", size=11)
        self.show_login_page()

    def clear_window(self):
        for widget in self.root.winfo_children():
            widget.destroy()

    def show_login_page(self):
        self.clear_window()
        main_frame = tk.Frame(self.root, bg="#f0f4f8")
        main_frame.place(relx=0.5, rely=0.5, anchor="center")
        logo_frame = tk.Frame(main_frame, bg="#4f46e5", width=80, height=80)
        logo_frame.pack(pady=(0, 20))
        tk.Label(logo_frame, text="‚ù§", font=("Arial", 40), bg="#4f46e5", fg="white").pack(expand=True)
        tk.Label(main_frame, text="AI Health Chatbot", font=self.title_font, bg="#f0f4f8", fg="#1e293b").pack(pady=(0, 5))
        tk.Label(main_frame, text="PubMed-Powered Research Assistant", font=self.normal_font, bg="#f0f4f8", fg="#64748b").pack(pady=(0, 30))
        card_frame = tk.Frame(main_frame, bg="white")
        card_frame.pack(padx=40, pady=20)
        card_inner = tk.Frame(card_frame, bg="white", padx=40, pady=40)
        card_inner.pack()
        tk.Label(card_inner, text="Welcome Back", font=self.heading_font, bg="white", fg="#1e293b").grid(row=0, column=0, columnspan=2, pady=(0, 25), sticky="w")
        tk.Label(card_inner, text="Username", font=self.normal_font, bg="white", fg="#475569").grid(row=1, column=0, columnspan=2, sticky="w", pady=(0, 8))
        self.username_entry = tk.Entry(card_inner, font=self.normal_font, width=30)
        self.username_entry.grid(row=2, column=0, columnspan=2, pady=(0, 20), ipady=8)
        self.username_entry.focus()
        tk.Label(card_inner, text="Password", font=self.normal_font, bg="white", fg="#475569").grid(row=3, column=0, columnspan=2, sticky="w", pady=(0, 8))
        self.password_entry = tk.Entry(card_inner, font=self.normal_font, width=30, show="‚óè")
        self.password_entry.grid(row=4, column=0, columnspan=2, pady=(0, 25), ipady=8)
        self.password_entry.bind('<Return>', lambda e: self.login())
        tk.Button(card_inner, text="Sign In", font=self.heading_font, bg="#4f46e5", fg="white", relief="flat", command=self.login, width=28, pady=10).grid(row=5, column=0, columnspan=2, pady=(0, 15))
        tk.Label(card_inner, text="Demo credentials: demo / demo123", font=("Helvetica", 9), bg="white", fg="#94a3b8").grid(row=6, column=0, columnspan=2)

    def login(self):
        username = self.username_entry.get().strip()
        password = self.password_entry.get().strip()
        if not username or not password:
            messagebox.showerror("Error", "Please enter both username and password")
            return
        user_id = authenticate_user(self.conn, username, password)
        if user_id:
            self.user_id = user_id
            self.username = username
            self.show_chatbot_page()
        else:
            messagebox.showerror("Login Failed", "Invalid username or password\n\nTry: demo / demo123")

    def show_chatbot_page(self):
        self.clear_window()
        header_frame = tk.Frame(self.root, bg="#4f46e5", height=70)
        header_frame.pack(fill="x")
        header_frame.pack_propagate(False)
        tk.Label(header_frame, text="üî¨ PubMed Research Assistant", font=self.title_font, bg="#4f46e5", fg="white").pack(side="left", padx=30, pady=15)
        tk.Label(header_frame, text=f"üë§ {self.username}", font=self.normal_font, bg="#4f46e5", fg="white").pack(side="right", padx=20)
        tk.Button(header_frame, text="Logout", font=self.normal_font, bg="#6366f1", fg="white", relief="flat", command=self.show_login_page, padx=15, pady=5).pack(side="right", padx=10)
        content_frame = tk.Frame(self.root, bg="#f0f4f8")
        content_frame.pack(fill="both", expand=True, padx=30, pady=20)
        search_frame = tk.Frame(content_frame, bg="white")
        search_frame.pack(fill="x", pady=(0, 20))
        search_inner = tk.Frame(search_frame, bg="white", padx=20, pady=20)
        search_inner.pack(fill="x")
        tk.Label(search_inner, text="Search PubMed", font=self.heading_font, bg="white", fg="#1e293b").pack(anchor="w", pady=(0, 10))
        search_container = tk.Frame(search_inner, bg="white")
        search_container.pack(fill="x")
        self.query_entry = tk.Entry(search_container, font=self.normal_font)
        self.query_entry.pack(side="left", fill="x", expand=True, ipady=8)
        self.query_entry.focus()
        self.query_entry.bind('<Return>', lambda e: self.search())
        tk.Button(search_container, text="üîç Search", font=self.normal_font, bg="#4f46e5", fg="white", relief="flat", command=self.search, padx=20, pady=9).pack(side="left", padx=(10, 0))
        result_count_frame = tk.Frame(search_inner, bg="white")
        result_count_frame.pack(fill="x", pady=(10, 0))
        tk.Label(result_count_frame, text="Results to display:", font=self.normal_font, bg="white", fg="#475569").pack(side="left", padx=(0, 10))
        self.result_count = tk.StringVar(value="20")
        result_options = ["10", "20", "50", "100", "200"]
        result_dropdown = ttk.Combobox(result_count_frame, textvariable=self.result_count, values=result_options, state="readonly", width=10, font=self.normal_font)
        result_dropdown.pack(side="left")
        results_frame = tk.Frame(content_frame, bg="white")
        results_frame.pack(fill="both", expand=True)
        results_header = tk.Frame(results_frame, bg="white", padx=20, pady=15)
        results_header.pack(fill="x")
        tk.Label(results_header, text="Research Articles", font=self.heading_font, bg="white", fg="#1e293b").pack(side="left")
        tk.Button(results_header, text="üìã View History", font=self.normal_font, bg="#e0e7ff", fg="#4f46e5", relief="flat", command=self.show_history, padx=15, pady=5).pack(side="right")
        self.results_text = scrolledtext.ScrolledText(results_frame, font=self.normal_font, wrap=tk.WORD, bg="#f9fafb", padx=20, pady=10)
        self.results_text.pack(fill="both", expand=True, padx=20, pady=(0, 20))
        self.results_text.insert("1.0", "Enter a disease, symptom, or medical term to search PubMed research articles.\n\nExample searches:\n‚Ä¢ diabetes type 2\n‚Ä¢ COVID-19 treatment\n‚Ä¢ alzheimer prevention\n\nSelect number of results (10-200) from dropdown above.")
        self.results_text.config(state="disabled")

    def search(self):
        query = self.query_entry.get().strip()
        if not query:
            messagebox.showwarning("Empty Query", "Please enter a search term")
            return
        try:
            max_results = int(self.result_count.get())
        except:
            max_results = 20
        self.results_text.config(state="normal")
        self.results_text.delete("1.0", tk.END)
        self.results_text.insert("1.0", f"üîé Searching PubMed for '{query}'...\n\n")
        self.results_text.config(state="disabled")
        self.root.update()
        papers = search_pubmed(query, max_results=max_results)
        save_history(self.conn, self.user_id, query, papers)
        self.results_text.config(state="normal")
        self.results_text.delete("1.0", tk.END)
        if not papers:
            self.results_text.insert("1.0", f"‚ùå No results found for '{query}'\n\nTry:\n‚Ä¢ Using different keywords\n‚Ä¢ Checking spelling\n‚Ä¢ Using medical terminology")
        else:
            self.results_text.insert("1.0", f"‚úÖ Found {len(papers)} articles for '{query}':\n\n")
            for i, paper in enumerate(papers, 1):
                self.results_text.insert(tk.END, f"‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ‚îÅ\n{i}. {paper['title']}\n\nüì∞ Journal: {paper['journal']}\nüë• Authors: {paper['authors']}\nüìÖ Published: {paper['pub_date']}\nüîó Link: {paper['url']}\n\n")
        self.results_text.config(state="disabled")

    def show_history(self):
        history = get_history(self.conn, self.user_id, limit=10)
        if not history:
            messagebox.showinfo("History", "No search history found")
            return
        history_window = tk.Toplevel(self.root)
        history_window.title("Search History")
        history_window.geometry("600x400")
        history_window.configure(bg="#f0f4f8")
        tk.Label(history_window, text="üìã Recent Searches", font=self.heading_font, bg="#f0f4f8", fg="#1e293b").pack(pady=20)
        tree_frame = tk.Frame(history_window, bg="white")
        tree_frame.pack(fill="both", expand=True, padx=20, pady=(0, 20))
        tree = ttk.Treeview(tree_frame, columns=("Query", "Date"), show="headings", height=15)
        tree.heading("Query", text="Search Query")
        tree.heading("Date", text="Date & Time")
        tree.column("Query", width=300)
        tree.column("Date", width=250)
        for query, timestamp in history:
            dt = datetime.fromisoformat(timestamp)
            tree.insert("", tk.END, values=(query, dt.strftime("%Y-%m-%d %H:%M:%S")))
        tree.pack(fill="both", expand=True)
        tk.Button(history_window, text="Close", font=self.normal_font, bg="#4f46e5", fg="white", relief="flat", command=history_window.destroy, padx=20, pady=8).pack(pady=(0, 20))


# if __name__ == "__main__":
#     root = tk.Tk()
#     app = PubMedChatbotApp(root)
#     root.mainloop()
```
